# Part1:
# D1: Select a folder / Create a folder with different file types ( at least 10 different ) and atleast 20-30 different files for each type

-> defining the function create_new_folder which takes the root path and name of the folder as the parameters.

-> os.path.join - this method is concatenates one or more paths with exactly one directory separator ('/') with each non - empty path except for the last non-empty part

-> os.path.exist - checks if the path exists or not, returns a boolean value

-> os.mkdir - makes a new directory with specified path

Procedure followed - 
1. taking root_path and name as arguments
2. create a new path using os.path.join
3. checking if the path does not exist then create a new directory using the new path
4. returns the new_path

```python
def create_new_file(root_path,name):
    new_path = os.path.join(root_path,name)
    
    if(not os.path.exists(new_path)):
        os.mkdir(new_path)
    
    return new_path
```

-> os.getcwd() - method by which we get the path of the current working directory

Procedure - 
1. get current path as root
2. create a new folder named D1 using root where the files will be intially added
3. create a new folder named D1_copy using root where the files will be copied

```python
root = os.getcwd()
d1 = create_new_file(root,'D1')
d1_copy = create_new_file(root,'D1_copy')
print(root)
print(d1)
print(d1_copy)
```

-> defining a dictionary with file types and extensions
procedure - 
1. Keys are the file type
2. values are the list of valid extesnions for that file


```python
types = {
    "HTML": [".html5", ".html", ".htm", ".xhtml"],
   "PYTHON": [".py"],
   "DOCUMENTS": [".oxps", ".epub", ".pages", ".docx", ".doc", ".fdf", ".ods",
   ".odt", ".pwi", ".xsn", ".xps", ".dotx", ".docm", ".dox",
   ".rvg", ".rtf", ".rtfd", ".wpd", ".xls", ".xlsx", ".ppt",
   ".pptx"],
   "PDF": [".pdf"],
   "XML": [".xml"],
   "EXE": [".exe"],
   "SHELL": [".sh"],
   "IMAGES": [".jpeg", ".jpg", ".tiff", ".gif", ".bmp", ".png", ".bpg", ".svg",
   ".heif", ".psd"],
   "VIDEOS": [".avi", ".flv", ".wmv", ".mov", ".mp4", ".webm", ".vob", ".mng",".qt", ".mpg", ".mpeg", ".3gp"],
   "ARCHIVES": [".a", ".ar", ".cpio", ".iso", ".tar", ".gz", ".rz", ".7z",
   ".dmg", ".rar", ".xar", ".zip"],
   "AUDIO": [".aac", ".aa", ".aac", ".dvf", ".m4a", ".m4b", ".m4p", ".mp3",
   ".msv", ".ogg", ".oga", ".raw", ".vox", ".wav", ".wma"],
   "PLAINTEXT": [".txt", ".in", ".out"]  
}
```

-> creating a new dictionary where for every value in 'types' is key in ext_types and the keys in 'types' are values in ext_type

```python
ext_type = dict()

for key in types.keys():
    for value in types[key]:
        ext_type[value] = key
        
print(ext_type)
```

# Part 2:
# Write a python programs which reads the folder D1 and based on the files creates the folder structure and copies / moves file from D1 to newly created folder structure.

-> creating a folder for every file type by calling the function

```python
for key in types.keys():
    create_new_file(d1_copy,key)
```

-> creating 30 files of each type

-> creating an empty dictionary with will store creation time for every file

```python
n = 30
file_ctime = dict()
```

-> random.choice() - selecting random values from the list of values in 'types'

-> os.path.join() - to concatenate one or more paths with name separated by '/'

-> os.path.ctime() - fetches the time of creation of the file in seconds for a specified path

-> time.ctime() - converts seconds to its string notation

Procedure - 
1. iterate through the keys of 'types'
2. creating 30 files for each key
3. extension fetches random valid extension from the list of values
4. name of the file is concatenated as - file type + nth number of the file + exttension
5. get the location of each file
6. write some content in each file
7. assign ctime to each file created

```python
for key in types:
    for i in range(1,n+1):
        extension = random.choice(types[key])
        file_name = key + str(i) + extension
        location=os.path.join(d1,file_name)
        
        with open(location,'w') as f:
            f.write("for PSC assignment ")
            f.write("This file "+str(key)+" has extension "+extension)
            f.write("and stands "+str(i)+"th number of the file.")
        
        file_ctime[file_name]=time.ctime(os.path.getctime(location))

print(file_ctime)
```

-> os.listdir() - listing of files and directories in the parent directory

-> os.path.splitext() - to file name name and its extension

-> shutil.move() - to move files/directories from one path to another

Procedure - 
1. fetch the list of all files 
2. find the name and extension of the file
3. compute original path
4. compute new path
5. move files from orginal path to new path

```python
file_directories = os.listdir(d1)

for i in file_directories:
    name,ext = os.path.splitext(i)
    type_file = ext_type[ext]
    move_to = os.path.join(d1_copy,type_file)
    move_from = os.path.join(d1,i)
    shutil.move(move_from,move_to)
```

# Part 3:
# creating csv file and storing it in parent folder

-> os.path.getctime - to get the creation time of a path

-> os.path.getmtime - to get the modification time of a path

-> to_csv - to convert a data structure to csv type of file

Procedure - 
1. declare a dictionary with column names
2. iterate through the keys of file_ctime
3. fetch the name and extension of the key
4. assign the original ctime in D1 folder to D1time
5. find the type of extension from the ext_type dictionary
6. compute the old path
7. compute the new path
8. compute ctime for new path
9. compute mtime for new path
10. assign the values with respective columns of 'data'
11. declare a DataFrame with column as that of keys of 'data'
12. convert it to csv file

```python
data = {'File type':[],'Extension':[],'Timestamp of creation in D1':[],'New creation time in D1 copy':[],
        'Timestamp of last modification':[],'Original file path':[],'New path after copying':[]}

for key in file_ctime.keys():
    name,ext = os.path.splitext(key)
    D1time = file_ctime[key]
    type_ext = ext_type[ext]
    old_path = os.path.join(d1,key)
    new_path = os.path.join(d1_copy,type_ext,key)
    copytime = os.path.getctime(new_path)
    mod_time = os.path.getmtime(new_path)
    
    data['File type'].append(name)
    data['Extension'].append(ext)
    data['Timestamp of creation in D1'].append(D1time)
    data['New creation time in D1 copy'].append(copytime)
    data['Timestamp of last modification'].append(mod_time)
    data['Original file path'].append(old_path)
    data['New path after copying'].append(new_path)
    
data_csv = pd.DataFrame(data,columns = ['File type','Extension','Timestamp of creation in D1','New creation time in D1 copy'
                                       'Timestamp of last modification','Original file path','New path after copying'])
data_csv.to_csv("18BCE101_CSV_log_file.csv")
data_csv.head(30)
```
